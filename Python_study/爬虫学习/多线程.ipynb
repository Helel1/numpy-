多线程 thread

```python
from threading import Thread
```

```python
def func():
    for i in range(100):
        print("func",i)

if __name__ == "__main__":
    t = Thread(target=func)
    t.start()
    for i in range(100):
        print("main",i)
```

继承Thread重写run函数

```python
class MyThread(Thread):
    def __init__(self):
        super().__init__()
        self.args = 1
    def run(self):
        for i in range(100):
            print(2,i)
        
if __name__ == "__main__":
    t = MyThread()
    t.start()
    for i in range(100):
        print(t.args,i)
```

## 多进程

```python
from multiprocessing import Process
```

```python
def func():
    for i in range(1000):
        print("func",i)

if __name__ == "__main__":
    p = Process(target=func)
    p.start()
    for i in range(1000):
        print("main",i)
```

```python
def func(name):
    for i in range(100):
        print(name,i)

if __name__ == "__main__":
    t1 = Thread(target=func,args=("1",))
    t1.start()
    t = Thread(target=func,args=("2",))
    t.start()
```

## 线程池

一次性开辟多个线程，用户提交给线程池

```python
from concurrent.futures import ThreadPoolExecutor as TPE
from concurrent.futures import ProcessPoolExecutor as PPE
```

```python
def func(name):
    for i in range(100):
        print(name,i)
```

```python
if __name__ == "__main__":
    with TPE(50) as t:
        for i in range(1000):
            t.submit(func,name=f"线程{i}")
    # 等待线程池中的执行完才会执行（守护）
    print("123")
```
