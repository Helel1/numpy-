```python
import requests
```

```python
url = f"http://www.xinfadi.com.cn/getPriceData.html"
data = {
        "limit": 20,
        "current": 1,
        "pubDateStartTime": '',
        "pubDateEndTime": '',
        "prodPcatid": '',
        "prodCatid": '',
        "prodName": '',
    }
resp = requests.post(url, data=data)
```

```python
dic = resp.json()
dic
```

```python
type(dic['list'])
```

```python
for i in dic['list']:
    prodName = i['prodName']
    print(prodName)
    lowPrice = i['lowPrice'] 
    print(lowPrice)
    highPrice = i['highPrice']  
    print(highPrice)
    avgPrice = i['avgPrice']
    print(avgPrice)
    pubDate = i['pubDate']
    print(pubDate)
```
