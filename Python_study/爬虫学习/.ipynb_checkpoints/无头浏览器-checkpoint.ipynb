# 无头浏览器

```python
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
chrome_options = webdriver.ChromeOptions()

chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-dev-shm-usage')


web = webdriver.Chrome(options=chrome_options)

```

```python
web.get('https://www.endata.com.cn/BoxOffice/BO/Year/index.html')
```

```python
select = web.find_element_by_xpath('//*[@id="OptionDate"]')
```

## 对元素进行包装，包装成下拉列表

```python
sel = Select(select)
```

## 让浏览器调整选项  

- sel.select_by_index
    通过索引切换
- sel.select_by_value
    通过值进行切换
- sel.select_by_visible_text
    通过可见的内容切换

```python
# sel.select_by_index(3) 切换到2018[2021.2020,2019,2018]
#sel.select_by_value("2017")
sel.select_by_visible_text("2014年")
```

```python
import csv

f = open('movie_top.csv',mode='a+',encoding='utf-8',newline="")
fw = csv.writer(f)
```

```python
for i in range(len(sel.options)):
    
    sel.select_by_index(i)
    time.sleep(10)
    for j in range(1,26):
        top = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[1]').text
        movie_name = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[2]/a/p').text
        leixing = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[3]').text
        zpf = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[4]').text
        price_avg = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[5]').text
        renci = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[6]').text
        cuntry = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[7]').text
        date = web.find_element_by_xpath(f'//*[@id="TableList"]/table/tbody/tr[{j}]/td[8]').text
        fw.writerow([top,movie_name,leixing,zpf,price_avg,renci,cuntry,date])
        print(([top,movie_name,leixing,zpf,price_avg,renci,cuntry,date]))
    print(f'第{i}个已完成')
f.close()
```
